# AI Notebook Import Smoke Test

Small fixture notebook for testing notebook-to-analysis imports.

Covers:
- local CSV import
- dataframe transform
- preview/output cell
- Google Sheets import
- local file export
- Google Sheets export


```python
import pandas as pd
import gspread
```

## Local file import


```python
df_sales = pd.read_csv("sample_sales.csv")
df_sales
```

## Simple transform


```python
df_sales_clean = df_sales.copy()
df_sales_clean["revenue"] = df_sales_clean["units"] * df_sales_clean["price"]
df_summary = (
    df_sales_clean.groupby("region", as_index=False)["revenue"]
    .sum()
    .sort_values("revenue", ascending=False)
)
```

```python
df_summary.head()
```

## Google Sheets import


```python
df_targets = pd.read_csv(
    "https://docs.google.com/spreadsheets/d/1ExampleSheetId1234567890/export?format=csv&gid=0"
)
df_targets
```

## Local export


```python
df_summary.to_csv("sales_summary.csv", index=False)
```

## Google Sheets export


```python
gc = gspread.service_account(filename="service-account.json")
sheet = gc.open_by_url(
    "https://docs.google.com/spreadsheets/d/1AnotherExampleSheetId0987654321/edit#gid=0"
).worksheet("Summary")
sheet.clear()
sheet.update([df_summary.columns.tolist()] + df_summary.values.tolist())
```
